```python
num_features=df.select_dtypes(include =['float64','int64']).columns.tolist()
num_features
```

```python

cat_features=df.select_dtypes(include ='object').columns.tolist()
cat_features
```

```python
num_features.remove('Id')
num_features.remove('SalePrice')
```

```python
num_features.remove('MSSubClass')
cat_features.append('MSSubClass')
```

```python
num_features.remove('OverallQual')
cat_features.append('OverallQual')
```

```python
num_features.remove('OverallCond')
cat_features.append('OverallCond')
```

```python
num_features.remove('MoSold')
cat_features.append('MoSold')
```

```python

cat_features.remove('PoolQC')
cat_features.remove('MiscFeature')
cat_features.remove('Fence')
cat_features.remove('FireplaceQu')
cat_features.remove('Alley')
# I dont use these feauterss, since there are a lot of values missing in this columns of the test dataframe
# I use all the other features because i want to use as many as possible
```

```python
#I delted outliers it improved my score signifcally, even if i loose data here
for col in num_features:
    col_zscore = col + '_zscore'
   #X[col_zscore] = (df[col] - df[col].mean())/df[col].std(ddof=0)
    df = df[(np.abs( (df[col] - df[col].mean())/df[col].std(ddof=0))< 3)]
    
#df.info()
```

```python
cat_features
```

```python
X = df.drop(['SalePrice'], axis=1)
y = df['SalePrice']
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=31)

print(X_train.shape)
print(X_test.shape)
```

```python


from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
num_pipeline = Pipeline(
    steps=[
        ('num_imputer', SimpleImputer()),
        ('scaler', StandardScaler()),
        ]
)
cat_pipeline = Pipeline(
    steps=[
        ('cat_imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown="ignore")),
    ]
)
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num_pipeline', num_pipeline, num_features),
        ('cat_pipeline', cat_pipeline, cat_features),
    ]
)
from sklearn.linear_model import LinearRegression

lin_reg_full_pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('lin_reg', LinearRegression()),
    ]
)
```

```python
lin_reg_full_pipeline.fit(X_train, y_train)
```

```python
housing_pred_lin= lin_reg_full_pipeline.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_test, housing_pred_lin) 
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg_full_pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
       ('sgd_reg', SGDRegressor()),
    ]
)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {
        
        'preprocessor__num_pipeline__num_imputer__strategy': ['mean', 'median'],
        'sgd_reg__max_iter': [200,300,500,1000,2000], 
        'sgd_reg__penalty': ['l2','l1','elasticnet'], 
        'sgd_reg__alpha': [1e-4, 1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3],
        'sgd_reg__eta0': [1e-1,1e-2,1e-3],
        
        
    }
]
# set up the grid search 
grid_search = GridSearchCV(sgd_reg_full_pipeline, param_grid, cv=5,
                          scoring='neg_mean_squared_error',
                          return_train_score=True)

# train the model using the full pipeline
grid_search.fit(X_train, y_train)
```

max_iter is the maximum number it goes over the training data and choses random data,  
tol is the tolerance when we stop iterating of the data , the tolerance is the norm of the gradients of our current theta,  
eta0 is the learning rate in the beginning

```python
sgd_reg_best = grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
housing_pred_sgd = sgd_reg_best.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_test, housing_pred_sgd) 
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python


y_pred2 = sgd_reg_best.predict(df_test)

sgd_reg_full_pipeline_submit = pd.DataFrame({
    'Id': df_test['Id'],
    'SalePrice': y_pred2
})
sgd_reg_full_pipeline_submit.to_csv('sg-reg-full-pipeline-submit.csv', index=False)
```
