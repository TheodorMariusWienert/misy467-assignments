## Task 1
There are 8 basic functions as described at http://mathonweb.com/help_ebook/html/functions_4.htm. Do the followings:
1. create 8 subplots (4x2 4 rows with two subplots each row) with figure size (15, 15)
2. plot a function (you choose) of each type in each subplot, add the function type as the title of the subplot, and the quadratic function has to be in the subplot in the second row and second column
3. use at least 4 colors, 4 different line styles, and 4 different markers for the 8 functions
4. change the x label to x and y label to the function
5. add legends for each subplot to show the functions, such as y = 3x +5

```python
fig,ax=plt.subplots(4,2,figsize=(15,15))
x=np.linspace(0,10,50)
ax[0,0].plot(x,np.log(x+1),linestyle='-',color='red',label='log(x+1)')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('log(x+1)')
ax[0,0].legend()

ax[0,1].plot(x,x**0.5,linestyle='--',color='green',label='$y=x^{0.5}$')
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('$x^{0.5}$')
ax[0,1].legend()

ax[1,0].plot(x,2*x,linestyle='-.',color='yellow',label='$y=2\cdot x$')
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('$2\cdot x$')
ax[1,0].legend()

ax[1,1].plot(x,x**2+2*x+1,linestyle=':',color='black',label='$y=x+2\cdot x+1$')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('$x+2\cdot x+1$')
ax[1,1].legend()

ax[2,0].plot(x,x**3+x*2+5,'*',label='$y=x+x\cdot2+5$')
ax[2,0].set_xlabel('x')
ax[2,0].set_ylabel('$x+x\cdot2+5$')
ax[2,0].legend()

ax[2,1].plot(x,(x**3+x*2+5)/(x**5+x*2+5),'o',label='$y=\dfrac {(x+x \cdot 2+5)}{(x+x \cdot 2+5)}$')
ax[2,1].set_xlabel('x')
ax[2,1].set_ylabel('$\dfrac {(x+x \cdot 2+5)}{(x+x \cdot 2+5)}$')
ax[2,1].legend()

ax[3,0].plot(x,3**x,'s',label='$y=3^{x}$')
ax[3,0].set_xlabel('x')
ax[3,0].set_ylabel('$3^{x}$')
ax[3,0].legend()

ax[3,1].plot(x,(np.sin(x)),'v',label='y=sin(x)')
ax[3,1].set_xlabel('x')
ax[3,1].set_ylabel('sin(x)')
ax[3,1].legend()

```

```python
fig=plt.figure(figsize=(15,15))
ax=plt.axes()
df_train['Sex'].hist(ax=ax)
survived_by_Sex=df_train.groupby('Sex')['Survived'].value_counts()
display(survived_by_Sex)
display(survived_by_Sex[0]/(survived_by_Sex[1]+survived_by_Sex[0]))
display(survived_by_Sex[3]/(survived_by_Sex[2]+survived_by_Sex[3]))
```
