```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train=pd.read_csv('Titanic_Data/titanic-train.csv')
```

```python
s_norm=df_train['Survived'].value_counts(normalize=True)
s_norm
```

```python
df_test=pd.read_csv('Titanic_Data/test.csv')
df_test.info()
```

```python
df_test['Survived']=0
df_test.head()
```

```python
df_submit_allzero=df_test[['PassengerId','Survived']]
df_submit_allzero.to_csv('Titanic_Data/titanic_submit_allzero.csv',index=False)
```

```python
sex_s_norm=df_train.groupby('Sex')['Survived'].value_counts(normalize=True)
sex_s_norm
```

```python
df_test2=pd.read_csv('Titanic_Data/test.csv')
```

```python
df_test2['Survived']=df_test2['Sex'].apply(lambda x: 0 if x=='male' else 1)
df_test2.head()
```

```python
df_submit_gender=df_test2[['PassengerId','Survived']]
df_submit_gender.to_csv('Titanic_Data/titanic_submit_gender.csv',index=False)
```

```python
pclass_s_norm=df_train.groupby('Pclass')['Survived'].value_counts(normalize=True)
pclass_s_norm
```

```python
df_test3=pd.read_csv('Titanic_Data/test.csv')
```

```python
df_test3['Survived']=df_test2['Pclass'].apply(lambda x: 1 if x==1  else 0)
df_test3.head(20)
```

```python
df_submit_Pclass=df_test3[['PassengerId','Survived']]
df_submit_Pclass.to_csv('Titanic_Data/titanic_submit_Pclass.csv',index=False)
```
